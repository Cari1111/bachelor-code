```python
def split_list(list, n):
    k, m = divmod(len(snapshots), n)
    return [list[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n)]
```

```python
n = pypsa.examples.scigrid_de()
```

```python
snapshots = n.snapshots
snapshot_splits = split_list(list=snapshots, n=5)
netw_splits = []
for snapshot_split in snapshot_splits:
    netw_copy = n.copy()
    netw_copy.set_snapshots(snapshots=snapshot_split)
    netw_copy.optimize(snapshot_split, solver_name="gurobi")
    netw_splits.append(netw_copy)
```

```python
feature_array = []
result_array = []

for netw in netw_splits:
    wind1_mean = np.mean(netw.generators_t.p_max_pu['7 Wind Onshore'])
    wind2_mean = np.mean(netw.generators_t.p_max_pu['271 Wind Offshore'])
    solar_mean = np.mean(netw.generators_t.p_max_pu['403_220kV Solar'])
    feature_array.append([wind1_mean, wind2_mean, solar_mean])

    capacity_mean = np.mean(netw.statistics()['Optimal Capacity'])
    result_array.append(capacity_mean)

print(feature_array)
print(result_array)
```
